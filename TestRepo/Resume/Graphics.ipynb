## Ley de Amdahl para escalabilidad fuerte
Para este apartado la formula es:

$Velocidad = \frac{1}{\frac{(s+p)}{N}}$ 

Donde s es la proporcin del tiempo serial, p es la porcion paralela y N es el nmero de procesadores.


```python
#Velocidad escalabilidad fuerte segn la ley de ahmdal

#Este es el tiempo completo (s+p)
Human = np.asarray(datae["Averaged"][:7])
Triticum = np.asarray(datae["Averaged"][9:16])
Pinus = np.asarray(datae["Averaged"][18:25])
#

Hs = 1/((Human-Human2).mean())
Ps = 1/((Pinus-Pinus2).mean())
Ts = 1/((Triticum-Triticum2).mean())

Hp = 1 - Hs
Pp = 1 - Ps
Tp = 1 - Ts

N = cores

# Aplicando la formula se SpeedUp de Ahmdal

SpeedUpH = 1/((Hs)+np.divide(Hp,N))
SpeedUpP = 1/((Ps)+np.divide(Pp,N))
SpeedUpT = 1/((Ts)+np.divide(Tp,N))


ScalH = Hs+np.multiply(Hp,N)
ScalP = Ps+np.multiply(Pp,N)
ScalT = Ts+np.multiply(Tp,N)

# Aplicando la formula: Haciendo esto se aplica la division elemento a elemento.
#serialhuman = (Human-Human2).mean()
#serialpinus = (Pinus-Pinus2).mean()
#serialtriticum = (Triticum-Triticum2).mean()
#velHuman = 1/((serialhuman)+Human2/cores)
#velTriticum = 1/((serialtriticum)+Triticum2/cores)
#velPinus = 1/((serialpinus)+Pinus2/cores)

#weakHuman = (Human-Human2)+Human2*cores
#weakTriticum = (Triticum-Triticum2)+Triticum2*cores
#weakPinus = (Pinus-Pinus2)+Pinus2*cores
```
